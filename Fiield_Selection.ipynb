```python
from tabulate import tabulate
from fractions import Fraction

def print_df(df):
    print(tabulate(df, headers = 'keys', tablefmt = 'psql'))
    
```

#### Extracting, Merging and Saving

```python
import pandas as pd
from tabulate import tabulate

excel_file = pd.ExcelFile('Field Selection\\MoraE21 DB.xlsx')
fields_df = pd.read_excel(excel_file, 'MoraE21 All')
results_df = pd.read_excel(excel_file, '1stSem')

grp = fields_df.groupby(['Field'])

#Mean vaules of SGPA for fields
mean_df = grp.mean('SGPA').sort_values('SGPA' , ascending = False)
#print('\n','Mean vaules of SGPA for fields: ')

#First and Last people's SGPA of a paticular filed
first = pd.Series(name = 'first')
last = pd.Series(name = 'last')
for name,group in grp:
    first[name] = group['SGPA'].max()
    last[name] = group['SGPA'].min()

mean_df['First'] = first
mean_df['Last'] = last

print_df(mean_df)

cols = list(results_df.columns)
results_df = results_df.drop(columns = cols[-4:])

results_df = results_df.drop(columns = ['Name'])
merged_df = fields_df.merge(results_df, on = 'ID')

#Formatting Columns
cols = list(merged_df.columns)
merged_df = merged_df[cols[:5] + [cols[-1]] + cols[7:14] + cols[5:7]]
cols = list(merged_df.columns)
cols[4] = 'SGPA'
cols[-3] = 'Exact SGPA'
merged_df.columns = cols

merged_df.to_csv('Field Selection\\Final_results.csv', index = False)
print('Saved as CSV')

merged_df

mech_df = merged_df[merged_df["Field"] == 'ME'][["ID","Name","UOM Mail"]]
mech_df
```

#### Analysis

```python
import pandas as pd

df = pd.read_csv('Field Selection\\Final_results.csv')
df = df.sort_values('Exact SGPA', ascending = False)

cols = list(df.columns)

new_df = df[['ID','Name','Sex','Field', 'SGPA']]
grp = new_df.groupby(['Field'])

#Males and Females
for name,group in grp:
    grp2 = group.groupby('Sex')
    if name != 'ME': continue
    for name2, group2 in grp2:
        if name2 == 'M': continue
        print(name,':')
        group2 = group2.sort_values('SGPA', ascending = False)
        group2 = group2.reset_index(drop = True)
        print(group2.shape[0])
        print(name2, print_df(group2))
        print_df(group2.describe())
```

```python
import pandas as pd

df = pd.read_csv('Field Selection\\Final_results.csv')

result = df.pivot_table(index='Field', columns='Sex', aggfunc='size', fill_value=0)
result = pd.DataFrame(result.to_records())
result.rename(columns={'F': 'Female Count', 'M': 'Male Count'}, inplace=True)
result['No of Males per Female'] = result['Male Count'] / result['Female Count']

result = result.sort_values('No of Males per Female', ascending = False)
print_df(result)
```
