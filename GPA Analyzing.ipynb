```python
import math
import tabula
import pandas as pd

cs_df = pd.read_csv('CSVs\\CS1033_results.csv')
mt_df = pd.read_csv('CSVs\\MT1023_results.csv')
ma_df = pd.read_csv('CSVs\\MA1014_results.csv')
ee_df = pd.read_csv('CSVs\\EE1040_results.csv')
ce_df = pd.read_csv('CSVs\\CE1023_results.csv')

grade_list = [cs_df,mt_df,ma_df,ee_df,ce_df]

def find_grade_count(df):
    return df.groupby(['Grade']).count()
 
filtered_data = []
for grade in grade_list:
    filtered_data.append(find_grade_count(grade))

final_df = pd.concat(filtered_data, axis = 1)
final_df.columns = ['CS', 'MT','MA','EE','CE']

#print(final_df)
grade_list = list(final_df.index)


grade_df = pd.read_csv('CSVs\\Grade_levels.csv')
grade_df = grade_df.drop(columns = ['Description','Benchmark Percentage'])
grades = grade_df.to_numpy().tolist()
print(grades)
avg_gpa_list = [0,0,0,0,0]
for row in final_df.itertuples():
    for grade in grades:
        if grade[0] == row[0]:
            for i in range(len(avg_gpa_list)):
                if not math.isnan(row[i + 1]):
                    avg_gpa_list[i] += float(grade[1]) * int(row[i + 1])
            #avg_gpa_list[0] += float(grade[1]) * row[1]
col_list = list(final_df.columns)
final_df.loc['Sum'] = final_df[col_list].sum(axis=0)
print(avg_gpa_list)
final_df.loc['Sum.'] = avg_gpa_list
final_df.loc['Avg.'] = final_df.loc['Sum.'] / final_df.loc['Sum']
col_list = list(final_df.columns)

print(final_df)
#print(cs_df.groupby('Grade').count().index)
'''
merged1_df = pd.merge(cs_df, mt_df, on="Index No.")
merged2_df = pd.merge(merged1_df, ma_df, on="Index No.")
merged3_df = pd.merge(merged2_df, ma_df, on="Index No.")
merged4_df = pd.merge(merged3_df, ce_df, on="Index No.")
print(merged4_df)
'''

#print(cs_df.join([mt_df,ma_df,ee_df,ce_df]))
#print(merged_df.loc[merged_df['Index No.'] == '210312V'])
```

```python
import pandas as pd
from tabulate import tabulate

df = pd.read_csv('CSVs\\selection_2020_results.csv')
df = df.drop(columns=['Depart'])

df = df.drop([164])
cols = list(df.columns)
bioE = df[cols[:4]]
bioE.dropna(inplace=True)

field_list = []
for index in range(0,21,4):
    field = df[cols[index:(index + 4)]]
    field.dropna(inplace = True)
    field.columns = ['Index','Name','Rank','SGPA']
    for row in field.iterrows():
        if not str(row[1].values[2]).isnumeric():
            field = field.drop([row[0]])
        #print(row[1].values[3])
        #print(type(row[1].values[2]))
    field_list.append(field)

gpa_series = pd.concat(field_list)
gpa_series = gpa_series.sort_values('SGPA' , ascending = False)
gpa_series = gpa_series.reset_index(drop = True)
gpa_series.SGPA = pd.to_numeric(gpa_series.SGPA)

print(gpa_series.count())
print('Mean SGPA :' , gpa_series.SGPA.mean()) , '\n'

print(tabulate(gpa_series.sample(n = 10), headers = 'keys' , tablefmt = 'rounded_outline'))

#print(df.iloc[164])
#print(selection_2020.loc[selection_2020['Index_Number'] == '210312V'])
```
